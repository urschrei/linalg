```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

coords = pd.read_pickle('data/coordinates.pickle')
fig = matplotlib.pyplot.gcf()
# this will save as 722px (max width for a Github README) when dpi is set to 100
fig.set_size_inches(7.22, 5.25)
orig = plt.scatter(
    coords['x_coordinates'],
    coords['y_coordinates'],
    color='#236b8e',
    edgecolor='k',
    lw=1,
    s=35,
    alpha=0.75)
upd = plt.scatter(
    coords['new_x'],
    coords['new_y'],
    marker='x',
    color='#FF7F00',
    lw=1,
    s=25)
plt.title("Imaged Points")
leg = plt.legend(
    (orig, upd),
    ('Original observations',
    'After least squares'),
    loc='lower left',
    scatterpoints=1,
    fontsize=8)
leg.get_frame().set_alpha(0.5)
plt.show()
# plt.savefig('results.png', dpi=100, alpha=True)
```
