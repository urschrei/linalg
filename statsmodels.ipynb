```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

A_matrix = pd.read_pickle('data/A_matrix.pickle')
b_vector = pd.read_pickle('data/b_vector.pickle')
similarity_A_matrix = pd.read_pickle('data/similarity_A_Matrix.pickle')
coords = pd.read_pickle('data/coordinates.pickle')
```

```python
affine_results = sm.OLS(b_vector, A_matrix).fit()
affine_results.summary()
```

```python
similarity_results = sm.OLS(b_vector, similarity_A_matrix).fit()
similarity_results.summary()
```

```python
coords['statsmodels_affine_x_residuals'] = pd.Series(list(affine_results.resid[0:54]))
coords['statsmodels_affine_y_residuals'] = pd.Series(list(affine_results.resid[54:108]))
coords['statsmodels_affine_x'] = coords['x_coordinates'] - coords['statsmodels_affine_x_residuals']
coords['statsmodels_affine_y'] = coords['y_coordinates'] - coords['statsmodels_affine_y_residuals']

coords['statsmodels_similarity_x_residuals'] = pd.Series(list(similarity_results.resid[0:54]))
coords['statsmodels_similarity_y_residuals'] = pd.Series(list(similarity_results.resid[54:108]))
coords['statsmodels_similarity_x'] = coords['x_coordinates'] - coords['statsmodels_similarity_x_residuals']
coords['statsmodels_similarity_y'] = coords['y_coordinates'] - coords['statsmodels_similarity_y_residuals']
```

```python
fig = matplotlib.pyplot.gcf()
# this will save as 722px (max width for a Github README) when dpi is set to 100
fig.set_size_inches(7.22, 5.25)
orig = plt.scatter(
    coords['x_coordinates'],
    coords['y_coordinates'],
    color='#236b8e',
    edgecolor='k',
    lw=1,
    s=35,
    alpha=0.75)
upd = plt.scatter(
    coords['new_x'],
    coords['new_y'],
    marker='+',
    color='w',
    lw=1,
    s=25)
af = plt.scatter(
    coords['statsmodels_affine_x'],
    coords['statsmodels_affine_y'],
    marker='x',
    color='#FF7F00',
    lw=1,
    s=25)
sm = plt.scatter(
    coords['statsmodels_similarity_x'],
    coords['statsmodels_similarity_y'],
    marker='^',
    color='r',
    lw=1,
    s=25,
    alpha=0.5,edgecolor='k')
plt.title("Imaged Points and Transformations")
leg = plt.legend(
    (orig, upd, af, sm),
    ('Original observations',
     'Affine transform using numpy.linalg.lstsq',
    'Affine transform using statsmodels.OLS',
    'Similarity transform using statsmodels.OLS'),
    loc='upper left',
    scatterpoints=1,
    fontsize=8)
leg.get_frame().set_alpha(0.5)
plt.tight_layout()
plt.savefig('results.png', dpi=100, alpha=True)
```
