```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

A_matrix = pd.read_pickle('data/A_matrix.pickle')
b_vector = pd.read_pickle('data/b_vector.pickle')
coords = pd.read_pickle('data/coordinates.pickle')
results = sm.OLS(b_vector, A_matrix).fit()
results.summary()
```

```python
coords['statsmodels_x_residuals'] = pd.Series(list(results.resid[0:54]))
coords['statsmodels_y_residuals'] = pd.Series(list(results.resid[54:108]))
coords['statsmodels_new_x'] = coords['x_coordinates'] - coords['statsmodels_x_residuals']
coords['statsmodels_new_y'] = coords['y_coordinates'] - coords['statsmodels_y_residuals']
```

```python
fig = matplotlib.pyplot.gcf()
# this will save as 722px (max width for a Github README) when dpi is set to 100
fig.set_size_inches(7.22, 5.25)
orig = plt.scatter(
    coords['x_coordinates'],
    coords['y_coordinates'],
    color='#236b8e',
    edgecolor='k',
    lw=1,
    s=35,
    alpha=0.75)
upd = plt.scatter(
    coords['new_x'],
    coords['new_y'],
    marker='+',
    color='w',
    lw=1,
    s=25)
sm = plt.scatter(
    coords['statsmodels_new_x'],
    coords['statsmodels_new_y'],
    marker='x',
    color='#FF7F00',
    lw=1,
    s=25)
plt.title("Imaged Points and Transformations")
leg = plt.legend(
    (orig, upd, sm),
    ('Original observations',
     'Affine transform using numpy.linalg.lstsq',
    'Affine transform using statsmodels.OLS'),
    loc='lower left',
    scatterpoints=1,
    fontsize=8)
leg.get_frame().set_alpha(0.5)
plt.tight_layout()
plt.savefig('results.png', dpi=100, alpha=True)
```
