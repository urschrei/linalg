The similarity matrix $A_{sim}$ is designed using the following equations:

\begin{equation}
f_i(a, b, \Delta{x}=ax_i + by_i + \Delta{x} \\\
f_i(a, b, \Delta{y}=bx_i + ay_i + \Delta{y}
\end{equation}

The affine matrix $A_{aff}$ is designed using the following equations:

\begin{equation}
f_i(a_0, a_1, a_2) = a_0 + a_1x_i + a_2y_i \\\
f_i(b_0, b_1, b_2) = a_0 + b_1x_i + b_2y_i
\end{equation}

The $b$ vector is identical for both transforms, and is formed of all the $x$ values, followed by all the $y$ values  
The covariance matrix $C_b$ is usually diagonal, but in this case is also a vector, of identical dimensions to $b$. It is formed of the square of the standard deviation of the $x$ values, for all $x$, followed by the square of the standard deviation of the $y$ values, for all $y$. It is converted into a diagonal weight matrix, **W**, in order to obtain the quality of the transformations.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm

affine_A_matrix = pd.read_pickle('data/A_matrix.pickle')
b_vector = pd.read_pickle('data/b_vector.pickle')
similarity_A_matrix = pd.read_pickle('data/similarity_A_Matrix.pickle')
coords = pd.read_pickle('data/coordinates.pickle')
covariance = pd.read_pickle('data/covariance_series.pickle')
```

```python
affine_results = sm.WLS(b_vector, affine_A_matrix, weights=1/covariance).fit()
affine_results.summary()
```

```python
similarity_results = sm.WLS(b_vector, similarity_A_matrix, weights=1/covariance).fit()
similarity_results.summary()
```

```python
coords['sm_affine_x_residuals'] = pd.Series(list(affine_results.resid[0:54]))
coords['sm_affine_y_residuals'] = pd.Series(list(affine_results.resid[54:108]))
coords['sm_affine_x'] = coords['x_coordinates'] - coords['sm_affine_x_residuals']
coords['sm_affine_y'] = coords['y_coordinates'] - coords['sm_affine_y_residuals']

coords['sm_similarity_x_residuals'] = pd.Series(list(similarity_results.resid[0:54]))
coords['sm_similarity_y_residuals'] = pd.Series(list(similarity_results.resid[54:108]))
coords['sm_similarity_x'] = coords['x_coordinates'] - coords['sm_similarity_x_residuals']
coords['sm_similarity_y'] = coords['y_coordinates'] - coords['sm_similarity_y_residuals']
```

```python
fig = matplotlib.pyplot.gcf()
# this will save as 722px (max width for a Github README) when dpi is set to 100
fig.set_size_inches(7.22, 5.25)
orig = plt.scatter(
    coords['x_coordinates'],
    coords['y_coordinates'],
    color='#33CCFF',
    edgecolor='#333333',
    marker='o',
    lw=1,
    s=80,
    alpha=1.0)
#upd = plt.scatter(
#    coords['new_x'],
#    coords['new_y'],
#    marker='o',
#    color='orange',
#    edgecolor='#666666',
#   lw=1,
#   s=25,
#    alpha=0.75)
af = plt.scatter(
    coords['sm_affine_x'],
    coords['sm_affine_y'],
    marker='o',
    color='#006600',
    edgecolor='#333333',
    lw=1,
    s=25,
    alpha=0.75)
sm = plt.scatter(
    coords['sm_similarity_x'],
    coords['sm_similarity_y'],
    marker='o',
    color='#CC00CC',
    edgecolor='#333333',
    lw=1,
    s=25,
    alpha=0.75)
plt.title("Imaged Points and Transformations")
leg = plt.legend(
    (orig, af, sm),
    ('Original observations',
    r'Affine transform using statsmodels.WLS - $\sigma$:%0.4f' % np.sqrt(affine_results.mse_resid),
    r'Similarity transform using statsmodels.WLS - $\sigma$:%0.4f' % np.sqrt(similarity_results.scale)),
    loc='upper left',
    scatterpoints=1,
    fontsize=8)
leg.get_frame().set_alpha(0.5)
plt.tight_layout()
plt.savefig('results.png', dpi=100, alpha=True)
```
